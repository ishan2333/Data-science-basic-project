```python
import json
```

```python
def load_data(filename):
    with open(filename , 'r') as f:
        data = json.load(f)
        
    return data
```

```python
data  = load_data('data.json')
```

```python
print(data)
```

```python
user_data = data['users']
```

```python
book_data = data['pages']
```

```python
print(user_data)
```

```python
for u in user_data:
    print(u)
```

```python
def find_user(id,data):
    for d in data:
        if d['id'] == id:
            return d['name']
```

```python
def find_book(id,data):
    for d in data:   
        if d['id'] == id:
            return d['name']
```

```python
find_book(104 , book_data)
```

```python
find_user(3,user_data)
```

```python
for user in user_data:
    print(user['liked_pages'])
    print(user['friends'])
    print(f"ID :- {user['id']} Name :- {user['name']}  Connected to {find_user(user['friends'],user_data)} , Liked Pages = {find_book(user['liked_pages'],book_data)} ")
```
